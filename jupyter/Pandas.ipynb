### Agrupar datos

```python
data_frame.groupby('ip').count().sample(10)
```

```python
ips_mas_consultadas = data_frame.groupby('ip').count().sort_values(by = 'metodo', ascending = False).head()
ips_mas_consultadas
```

```python
ips_mas_consultadas = ips_mas_consultadas[['fecha']]
ips_mas_consultadas
```

```python
data_frame = pd.read_csv('access.csv', delimiter = ',')
```
